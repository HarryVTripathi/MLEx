```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
```

```python
x = np.arange(0,4*np.pi,0.05)
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
fig = plt.figure(figsize=(16,12))
ax = fig.gca(projection='3d')

# Make data.
X = np.arange(-1.6, 1.6, 0.01)
Y = np.arange(-1.6, 1.6, 0.01)
X, Y = np.meshgrid(X, Y)
Z = np.exp(X**2 + Y**2)
#Z = np.sin(R)

# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm, linewidth=0, antialiased=False)

# Customize the z axis.
ax.set_zlim(-1, 10)
ax.zaxis.set_major_locator(LinearLocator(5))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

```python
X = np.arange(-10, 10, 1)
Y = np.arange(-10, 10, 1)
U, V = np.meshgrid(X, Y)

fig, ax = plt.subplots()
q = ax.quiver(X, Y, U, V)
ax.quiverkey(q, X=0.3, Y=1.1, U=10, label='Quiver key, length = 10', labelpos='E')

plt.show()
```

```python
x, y = np.arange(-10, 11, 1.36), np.arange(-10, 11, 1.36)
U, V = np.meshgrid(x, y)

X = V**3 -9*V
Y = U**3 -9*V

fig1, ax1 = plt.subplots()
Q = ax1.quiver(U, V, X, Y, units='width', angles='xy', scale=10000)
qk = ax1.quiverkey(Q, 1, 1, 2, r'$2 \frac{m}{s}$', labelpos='E', coordinates='figure')

plt.show()
```

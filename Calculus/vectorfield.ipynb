```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
x, y = np.mgrid[-5:5, -5:5]
X = y**3 - 9*y
Y = x**3 - 9*x

print(x)
print(y)

print(X)
print(Y)

fig, (ax_l, ax_r) = plt.subplots(1, 2, figsize=(16, 8))
ax_r.quiver(x, y, X, Y, scale=100, scale_units='inches')
ax_l.quiver(x, y, 2*X, 2*Y, scale=100, scale_units='inches')
ax_l.set_title('2x')
ax_r.set_title('1x')
```

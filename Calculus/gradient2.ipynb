#### Defining the ```function``` which returns the 2-D function as SymPy expression

````python
def fxy(x, y):
    return x**2 + y**2

def fxy2(x, y):
    return x*y
````

<br>
#### Creating meshgrids

````python
x, y, z, t = symbols('x y z t')
X = np.arange(-10, 11, 0.6)
Y = np.arange(-10, 11, 0.6)

U, V = np.meshgrid(X, Y)

# Cartesian product of X and Y
I = np.array(np.meshgrid(X, Y)).T.reshape(-1, 2)
# print(I[:, 0])
# print(I[:, 1])
````

#### Obtaining gradient - vector valued function

````python

## SimPy expressions
delx = diff(fxy2(x, y), x)
dely = diff(fxy2(x, y), y)


## Converting SymPy expressions into numeric function
fx = lambdify([x, y], delx, 'numpy')
fy = lambdify([x, y], dely, 'numpy')

# Vectors obtained after passing all possible (x X y) values into functions fx and fy
delfx = fx(I[:, 0], I[:, 1])
delfy = fy(I[:, 0], I[:, 1])

gradient = np.array([delfx, delfy])
````

````python
delfx = delfx.reshape(U.shape)
delfy = delfy.reshape(U.shape)
````

````python
fig, ax_r = plt.subplots(1, 1, figsize=(12, 12))
ax_r.quiver(V, U, delfx, delfy, scale=60, scale_units='inches')
ax_r.set_title('1x')
````

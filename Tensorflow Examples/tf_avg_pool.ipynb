```
michele = img.imread('/content/gdrive/My Drive/micheleboyd_70699459_106150380708098_6747831498182051626_n.jpg')
```

```
plt.imshow(michele)
```

```
michele = tf.dtypes.cast(x=michele, dtype='float32')

michele_tensor = tf.reshape(
    tensor=michele,
    shape=(1, 855, 1080, 3),
    name="michele_tensor",

)
```

```
pool2 = tf.nn.avg_pool2d(
    value=michele_tensor,
    ksize=[1, 2, 2, 1],
    strides=[1, 2, 2, 1],
    padding="SAME"
)

pooled_michele = tf.reshape(tensor=pool2, shape=(428, 540, 3))
```

```
with tf.Session() as sess:
  # print(sess.run(a))
  # print(sess.run(pool1))
  plt.imshow(sess.run(pooled_michele))
```

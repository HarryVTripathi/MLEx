```python
from matplotlib import pyplot as plt
```

#### A simple plot: Use the plot() method and pass X-axis and Y-axis values in it

```python
x = [1,2,4]
y = [4,2,1]
plt.plot(x,y)
```

#### Adding labels axes and title to the graph

```python
plt.title('Info')
plt.ylabel('Y-axis')
plt.xlabel('X-axis')
plt.plot(x,y)
plt.show()
```

#### Adding style to graph

```python
from matplotlib import style
style.use('ggplot')
```

```python
x2 = [2,3,5]
y2 = [2,3,5]
```

```python
plt.plot(x,y,'g',label='line1',linewidth=5)
plt.plot(x2,y2,'c',label='line2',linewidth=5)
plt.show()
```

#### Adding a legend to the graph

```python
plt.plot(x,y,'g',label='line1',linewidth=5)
plt.plot(x2,y2,'c',label='line2',linewidth=5)
plt.legend()
plt.title('Epic info')
plt.grid(True,color='k')
plt.show()
```

## Bar plots

###### Bar plots are used to compare between different groups

```python
x3 = [1,3,5,7,9]
x4 = [2,4,6,8,10]
y3 = [1,2,2,2,3]
y4 = [2,3,4,4,4]
plt.bar(x3,y3,label='one')
plt.bar(x4,y4,label='two',color='g')
plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
```

## Histogram

#### A histogram is used to represent quantitative variables while a bar graph is used for representing categorical variables

```python
import numpy as np
population_ages = [22,52,62,45,21,22,34,42,42,4,99,102,110,120,121,122,130,111,115,112,80,75,65,54,44,43,42,48]
bins = np.linspace(0,130,14)
plt.hist(population_ages,bins,histtype='bar',rwidth=0.9)
plt.show()
```

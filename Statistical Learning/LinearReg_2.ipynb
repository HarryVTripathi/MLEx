```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
X = np.random.randint(200, 800, 50)
Y = 20 + 2 * X
Y_obs = Y + np.random.randint(-240, 240, len(X))
```

```python
X_range = np.arange(200, 800, 0.4)
Y_range = 20 + 2 * X_range

plt.plot(X_range, Y_range)
plt.scatter(X, Y_obs, color='r', s=12)
plt.show()
```

```python
X = X.reshape(len(X),1)
X_test = X_range.reshape((len(X_range),1))

reg = LinearRegression()
reg = reg.fit(X, Y_obs)
LinearRegression(copy_X=True, fit_intercept=True, n_jobs=None, normalize=False)

Y_predicted = reg.predict(X_test)

print(reg.coef_[0])
print(reg.intercept_)
```

```python
plt.plot(X_range, Y_range)
plt.plot(X_test, Y_predicted, 'y')
plt.scatter(X, Y_obs, color='r', s=12)
plt.show()
```
